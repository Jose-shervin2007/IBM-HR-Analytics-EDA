# 5. Univariate Analysis

Univariate analysis examines one variable at a time. Various charts are created to understand the distribution of employee attributes such as age, gender, department, education level, and attrition status.

```python
plt.figure(figsize=(8,5))
sns.set_style("whitegrid")
```

## 5.1 Employee Attrition Distribution

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Attrition', data=df)
plt.title("Employee Attrition Count")
plt.xlabel("Attrition")
plt.ylabel("Number of Employees")
plt.show()
```

### Observation

Most employees stayed with the company, while a smaller percentage left. This indicates that the overall attrition rate is relatively low.

## 5.2 Gender Distribution

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Gender', data=df)
plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Number of Employees")
plt.show()
```

### Observation

The workforce consists of both male and female employees, with male employees slightly outnumbering female employees.

## 5.3 Department Distribution

```python
plt.figure(figsize=(8,5))
sns.countplot(x='Department', data=df)
plt.title("Employees by Department")
plt.xlabel("Department")
plt.ylabel("Number of Employees")
plt.xticks(rotation=15)
plt.show()
```

### Observation

The Research & Development department has the largest workforce, followed by Sales, while Human Resources has the fewest employees.

## 5.4 Education Level Distribution

```python
plt.figure(figsize=(7,4))
sns.countplot(x='Education', data=df)
plt.title("Education Level Distribution")
plt.xlabel("Education Level")
plt.ylabel("Number of Employees")
plt.show()
```

### Observation

Most employees have a Bachelor's or College-level education, indicating a well-qualified workforce.

# 6. Bivariate Analysis

Bivariate analysis examines the relationship between two variables. This analysis helps identify factors that may influence employee attrition.

## 6.1 Overtime vs Attrition

```python
plt.figure(figsize=(6,4))
sns.countplot(x='OverTime', hue='Attrition', data=df)
plt.title("Overtime vs Employee Attrition")
plt.xlabel("OverTime")
plt.ylabel("Employee Count")
plt.show()
```

### Observation

Employees working overtime have a higher attrition rate than employees who do not work overtime.

## 6.2 Department vs Attrition

```python
plt.figure(figsize=(8,5))
sns.countplot(x='Department', hue='Attrition', data=df)
plt.title("Department vs Attrition")
plt.xlabel("Department")
plt.ylabel("Employee Count")
plt.xticks(rotation=10)
plt.show()
```

### Observation

Employee attrition is observed across all departments, with Research & Development and Sales having higher numbers due to their larger workforce.

## 6.3 Job Role vs Attrition

```python
plt.figure(figsize=(12,5))
sns.countplot(x='JobRole', hue='Attrition', data=df)
plt.title("Job Role vs Attrition")
plt.xlabel("Job Role")
plt.ylabel("Employee Count")
plt.xticks(rotation=45)
plt.show()
```

### Observation

Attrition varies across different job roles, indicating that certain positions experience higher employee turnover.

## 6.4 Gender vs Attrition

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Gender', hue='Attrition', data=df)
plt.title("Gender vs Attrition")
plt.xlabel("Gender")
plt.ylabel("Employee Count")
plt.show()
```

### Observation

Both male and female employees experience attrition, suggesting that gender alone is not a major influencing factor.

## 6.5 Education Level vs Attrition

```python
plt.figure(figsize=(8,4))
sns.countplot(x='Education', hue='Attrition', data=df)
plt.title("Education Level vs Attrition")
plt.xlabel("Education Level")
plt.ylabel("Employee Count")
plt.show()
```

### Observation

Employees from different education levels show varying attrition rates, indicating that education may have some influence on employee retention.

# 7. Correlation Analysis

Correlation analysis is used to measure the relationship between numerical variables. It helps identify which features are positively or negatively related.

```python
numeric_df = df.select_dtypes(include=['int64', 'float64'])
numeric_df.head()
```

```python
correlation_matrix = numeric_df.corr()

correlation_matrix
```

```python
plt.figure(figsize=(18,12))

sns.heatmap(correlation_matrix,
            annot=True,
            cmap="coolwarm",
            fmt=".2f")

plt.title("Correlation Heatmap")

plt.show()
```

### Observation

- The heatmap displays the correlation between all numerical variables.
- Dark red colors indicate strong positive correlations.
- Dark blue colors indicate strong negative correlations.
- Most variables show weak to moderate correlation, indicating that employee attrition is influenced by multiple factors rather than a single feature.

# 8. Outlier Detection

Box plots are used to identify outliers in numerical variables. Outliers represent unusually high or low values that may require further investigation.


## 8.1 Age

```python
plt.figure(figsize=(6,4))

sns.boxplot(y=df["Age"])

plt.title("Box Plot of Age")

plt.show()
```

### Observation

Most employees fall within the expected age range, with only a few extreme values.

## 8.2 Monthly Income

```python
plt.figure(figsize=(6,4))

sns.boxplot(y=df["MonthlyIncome"])

plt.title("Box Plot of Monthly Income")

plt.show()
```

### Observation

Monthly income contains several outliers, indicating that some employees earn significantly more than others.

## 8.3 Total Working Years

```python
plt.figure(figsize=(6,4))

sns.boxplot(y=df["TotalWorkingYears"])

plt.title("Box Plot of Total Working Years")

plt.show()
```

### Observation

Most employees have moderate work experience, while a small number have significantly higher experience.
